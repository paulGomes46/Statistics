```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# =========================
# DATA
# =========================
data = {
    "HL": [41018539.54, 77797013.95, 8679668.24, 322631906.14, 60425458.44, 633822783.03],
    "IL": [38399185.37, 72080020.80, 11345674.11, 253254831.87, 67560120.02, 528487930.66],
    "HO": [802287.88, 1360972.95, 6042336.60, 6407446.84, 5666320.87, 51377595.28],
    "IO": [1002256.55, 1720716.61, 3614334.41, 5879385.90, 3608748.66, 33475448.71]
}

genes = [
    "Astragalin",
    "Kaempferol",
    "Quercetin",
    "Nicotiflorin",
    "Isoquercitrin",
    "Rutin"
]

df = pd.DataFrame(data, index=genes)

# =========================
# TRANSFORMATION
# =========================
# log2 transform (important vu les différences énormes)
df_log = np.log2(df + 1)

# option: scaling par ligne (recommandé pour pattern)
df_scaled = df_log.sub(df_log.mean(axis=1), axis=0)
df_scaled = df_scaled.div(df_log.std(axis=1), axis=0)

# =========================
# PLOT
# =========================
plt.figure(figsize=(5, 1.8))  # 👈 clé principale (plus large, moins haut)

sns.heatmap(
    df_scaled,
    cmap="RdBu_r",
    center=0,
    linewidths=0.5,
    linecolor="black",
    cbar_kws={
        "label": "Z-score",
        "shrink": 0.8
    }
)

plt.xticks(rotation=0)
plt.yticks(rotation=0)

plt.title("Monoterpene-related genes", fontsize=8)

plt.tight_layout()

plt.savefig("C:/Users/tamer/Documents/PhD/ML/metabolites_HM/fdp_metabolite.pdf")

plt.show()
```
