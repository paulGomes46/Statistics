```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = "C:/Users/tamer/Documents/PhD/ML/Markers/markers.xlsx"
df = pd.read_excel(data, sheet_name='Sheet8')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Style seaborn propre
sns.set_style("white")
sns.set_context("talk")

fig, ax = plt.subplots(figsize=(14,6))

# Couleurs sobres (colorblind-friendly)
palette = {
    "Constitutif": "#1f3b73",
    "I-1": "#d95f02",
    "I-2": "#757575"
}

x = np.arange(len(df))
width = 0.7

# Empilement
ax.bar(x, df["Constitutif"], width, 
       color=palette["Constitutif"], label="Constitutif")

ax.bar(x, df["I-1"], width, 
       bottom=df["Constitutif"], 
       color=palette["I-1"], label="I-1")

ax.bar(x, df["I-2"], width, 
       bottom=df["Constitutif"] + df["I-1"], 
       color=palette["I-2"], label="I-2")

# Labels
ax.set_xticks(x)
ax.set_xticklabels(df["Metabolite_family"], rotation=45, ha="right")
ax.set_ylabel("Number of metabolites")

# Nettoyage axes
sns.despine()
ax.legend(frameon=False)

plt.tight_layout()

plt.savefig("C:/Users/tamer/Documents/PhD/ML/Markers/bar_chart.pdf", format="pdf", bbox_inches="tight")   
plt.savefig("C:/Users/tamer/Documents/PhD/ML/Markers/bar_chart.png", dpi=600, bbox_inches="tight")  


plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set_style("white")
sns.set_context("talk")

fig, ax = plt.subplots(figsize=(14,6))

palette = {
    "Constitutif": "#1f3b73",
    "I-1": "#d95f02",
    "I-2": "#757575"
}

x = np.arange(len(df))
width = 0.7

# Empilement avec bordures
ax.bar(x, df["Constitutif"], width,
       color=palette["Constitutif"],
       edgecolor="black",
       linewidth=0.8,
       label="Constitutif")

ax.bar(x, df["I-1"], width,
       bottom=df["Constitutif"],
       color=palette["I-1"],
       edgecolor="black",
       linewidth=0.8,
       label="I-1")

ax.bar(x, df["I-2"], width,
       bottom=df["Constitutif"] + df["I-1"],
       color=palette["I-2"],
       edgecolor="black",
       linewidth=0.8,
       label="I-2")

ax.set_xticks(x)
ax.set_xticklabels(df["Metabolite_family"], rotation=45, ha="right")
ax.set_ylabel("Number of metabolites")

sns.despine()
ax.legend(frameon=False)

plt.tight_layout()

plt.savefig("C:/Users/tamer/Documents/PhD/ML/Markers/bar_chart.pdf",
            format="pdf", bbox_inches="tight")

plt.savefig("C:/Users/tamer/Documents/PhD/ML/Markers/bar_chart.png",
            dpi=600, bbox_inches="tight")

plt.show()
```
