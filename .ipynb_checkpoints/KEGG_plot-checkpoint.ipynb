```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# -------------------------
# LOAD DATA
# -------------------------
df = pd.read_excel(
    "C:/Users/tamer/Documents/PhD/ML/KEGG_enrichment/bubble_plot.xlsx",
    sheet_name = 'v2'
)

# garder colonnes utiles
meta_cols = ["Comparison", "Class"]
value_cols = [c for c in df.columns if c not in meta_cols]

# matrice
data = df[value_cols].values

# -------------------------
# GLOBAL STYLE (journal)
# -------------------------
plt.rcParams.update({
    "font.family": "Arial",
    "font.size": 8,
    "axes.titlesize": 10,
    "axes.labelsize": 9,
    "xtick.labelsize": 7,
    "ytick.labelsize": 7
})

# -------------------------
# FIGURE (double column)
# -------------------------
fig, ax = plt.subplots(figsize=(7.2, 5))  # ~18 cm

im = ax.imshow(data.T, aspect="auto")

# -------------------------
# Y axis (pathways)
# -------------------------
ax.set_yticks(range(len(value_cols)))
ax.set_yticklabels(value_cols)

# -------------------------
# X axis (comparison - level 2)
# -------------------------
ax.set_xticks(range(len(df)))
ax.set_xticklabels(df["Comparison"], rotation=45, ha="right")

# -------------------------
# TOP AXIS (class - level 1)
# -------------------------
ax_top = ax.twiny()
ax_top.set_xlim(ax.get_xlim())

# trouver centres des groupes
group_centers = []
group_labels = []

for cls in df["Class"].unique():
    idx = np.where(df["Class"] == cls)[0]
    center = np.mean(idx)
    group_centers.append(center)
    group_labels.append(cls)

ax_top.set_xticks(group_centers)
ax_top.set_xticklabels(group_labels)

ax_top.tick_params(axis='x', length=0, pad=2)

# -------------------------
# GROUP BARS (important)
# -------------------------
for cls in df["Class"].unique():
    idx = np.where(df["Class"] == cls)[0]
    start = idx[0] - 0.5
    end = idx[-1] + 0.5

    ax.hlines(
        y=-0.8,
        xmin=start,
        xmax=end,
        color="black",
        linewidth=1
    )

# -------------------------
# COLORBAR
# -------------------------
cbar = plt.colorbar(im, ax=ax, fraction=0.025, pad=0.02)
cbar.set_label("-log10(p-value)")

# -------------------------
# TITLE
# -------------------------
ax.set_title("KEGG pathway enrichment across metabolomic comparisons", pad=6)

# -------------------------
# LAYOUT
# -------------------------
plt.tight_layout()

plt.savefig(
    "C:/Users/tamer/Documents/PhD/ML/KEGG_enrichment/heatmap_double_column.tif",
    dpi=600,
    bbox_inches="tight"
)

plt.savefig(
    "C:/Users/tamer/Documents/PhD/ML/KEGG_enrichment/heatmap_double_column.pdf",
    bbox_inches="tight"
)

plt.show()
```
