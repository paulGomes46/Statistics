```python
labels = [
'Phenylpropanoids',
'Terpenoids',
'Alkaloids',
'Quinone',
'Benzenoids',
'Organic acids',
'Flavonoids',
'Cinnamic acids<br>and derivatives',
'Coumarins'
]


parents = [
' ',
' ',
' ',
' ',
' ',
' ',
'Phenylpropanoids',
'Phenylpropanoids',
'Phenylpropanoids',
]


values = [
'9',
'5',
'4',
'2',
'1',
'1',
'6',
'1',
'2'
]


fig =go.Figure(go.Sunburst(
    labels=labels,
    parents=parents,
    values=values,
    branchvalues="total",
    marker=dict(line=dict(color="black", width=2)),  # contour noir
    textinfo="label+value"  # ajouter valeurs
))

# Update layout for tight margin
fig.update_layout(margin = dict(t=0, l=0, r=0, b=0))

fig.update_layout(
    width=900,   # adjust width as needed
    height=900,   # adjust height as needed
    sunburstcolorway=px.colors.qualitative.Safe,
    extendsunburstcolors=True
)

fig.show()
```
