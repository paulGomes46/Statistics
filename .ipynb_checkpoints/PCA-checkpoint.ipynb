```python
def encodage(df):
    code = {
    'LP' : 'Lemon inoculated',
    'SP' : 'Orange inoculated',
    'LN' : 'Lemon control',
    'SN' : 'Orange control'
}
# Appliquer ce dictionnaire aux colonnes du dataset, avec la fonction map
    for col in df.select_dtypes('object'):
        df[col] = df[col].map(code)

    return df


def preprocessing(df):
    df = encodage(df)

    X = df.drop(['Class'], axis = 1)
    y = df['Class']

    # compter le nombre d'échantillons restants dans le dataset après avoir été inputé
    print(y.value_counts())

    return X, y
```

```python
X, y = preprocessing(df)
# Principalement encoder les classes
```

```python
import numpy as np
from matplotlib.patches import Ellipse

def confidence_ellipse(x, y, ax, n_std=2.0, **kwargs):
    x = np.asarray(x)
    y = np.asarray(y)

    if x.size < 2:
        return  # not enough points

    cov = np.cov(x, y)
    vals, vecs = np.linalg.eigh(cov)
    # sort eigenvalues / vectors
    order = vals.argsort()[::-1]
    vals, vecs = vals[order], vecs[:, order]

    # angle of first eigenvector
    theta = np.degrees(np.arctan2(*vecs[:, 0][::-1]))
    # width & height of ellipse = 2 * n_std * sqrt(eigenvalues)
    width, height = 2 * n_std * np.sqrt(vals)

    mean_x, mean_y = x.mean(), y.mean()
    ell = Ellipse((mean_x, mean_y),
                  width=width,
                  height=height,
                  angle=theta,
                  fill=False,
                  **kwargs)
    ax.add_patch(ell)
    return ell
```

```python
log2_transformer = FunctionTransformer(lambda x: np.log2(x + 1))
model = make_pipeline(log2_transformer, StandardScaler(), PCA(n_components=2))

X_reduced = model.fit_transform(X)
pca = model.named_steps['pca']

fig, ax = plt.subplots(figsize=(9, 6))

classes = y.unique()
colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red']

for cls, color in zip(classes, colors):
    mask = (y == cls)
    xg = X_reduced[mask, 0]
    yg = X_reduced[mask, 1]

    # points
    ax.scatter(
        xg, yg,
        color=color,
        label=cls,
        s=50,
        edgecolor='black',
        alpha=0.7
    )

    # ellipse around the group
    confidence_ellipse(
        xg, yg, ax,
        n_std=2.0,             
        edgecolor=color,
        linewidth=1,
        alpha=0.7
    )

plt.xlabel(f"PC1 = {pca.explained_variance_ratio_[0]*100:.2f}%", fontsize = 12)
plt.ylabel(f"PC2 = {pca.explained_variance_ratio_[1]*100:.2f}%", fontsize = 12)

ax.grid(True, which='both', linestyle='--', alpha=0.5)

leg = ax.legend(title="Groups", bbox_to_anchor=(1.05, 1), loc='upper left')
leg.get_title().set_fontweight('bold')

plt.tight_layout()

plt.savefig("C:/Users/tamer/Documents/PhD/ML/PCA/PCA_plot.pdf", format="pdf", bbox_inches="tight")   
plt.savefig("C:/Users/tamer/Documents/PhD/ML/PCA/PCA_plot.png", dpi=600, bbox_inches="tight")        

plt.show()
```
