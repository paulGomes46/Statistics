```python
def encodage(df):
    code = {
    'LP' : 1,
    'SP' : 2,
    'LN' : 3,
    'SN' : 4
}
# Appliquer ce dictionnaire aux colonnes du dataset, avec la fonction map
    for col in df.select_dtypes('object'):
        df[col] = df[col].map(code)

    return df


def preprocessing(df):
    df = encodage(df)

    X = df.drop(['Class'], axis = 1)
    y = df['Class']

    # compter le nombre d'échantillons restants dans le dataset après avoir été inputé
    print(y.value_counts())

    return X, y
```

```python
X_total, y_total = preprocessing(df)
```

```python
log2_transformer = FunctionTransformer(lambda x: np.log2(x + 1))
#model = make_pipeline(log2_transformer, StandardScaler(), SelectFpr(f_classif, alpha=0.05), LogisticRegression(penalty="elasticnet", l1_ratio=0.9, solver="saga", random_state=0))
model = make_pipeline(log2_transformer, StandardScaler(), SelectFpr(f_classif, alpha=0.05), LogisticRegression(penalty="l1", C=0.3, solver="saga", random_state=0))

model.fit(X_total, y_total)

#SelectFromModel(LogisticRegression(penalty="elasticnet", solver="saga", l1_ratio=0.5, random_state=0), threshold=None), 
```
