```python
import pandas as pd
import numpy as np
```

```python
dataset = "C:/Users/tamer/Documents/PhD/ML/HLB_phenotype/Phenotype.xlsx"
df = pd.read_excel(dataset, sheet_name = 'Leaf')
```

```python
df.head()
```

```python
from scipy.stats import mannwhitneyu

lemon = df[df.Genotype=="IL"]["Value"]
orange = df[df.Genotype=="IO"]["Value"]

stat, p = mannwhitneyu(lemon, orange)

print(p)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# ---------------- style publication ----------------
plt.rcParams.update({
    "font.family": "Arial",
    "font.size": 7,
    "axes.linewidth": 0.8,
    "xtick.major.width": 0.8,
    "ytick.major.width": 0.8
})

palette = {"IL": "#4c73b2",
           "IO": "#de8552"}

order = ["IL","IO"]

# ---------------- figure size ----------------
fig, ax = plt.subplots(figsize=(1.89, 2.36))

# ---------------- draw boxplots separately ----------------
for genotype in order:

    subset = df[df["Genotype"] == genotype]

    sns.boxplot(
        data=subset,
        x="Genotype",
        y="Value",
        order=order,
        width=0.5,
        showfliers=False,
        boxprops=dict(facecolor="white",
                      edgecolor=palette[genotype],
                      linewidth=1),
        whiskerprops=dict(color=palette[genotype], linewidth=1),
        capprops=dict(color=palette[genotype], linewidth=1),
        medianprops=dict(color=palette[genotype], linewidth=1),
        ax=ax
    )

# ---------------- scatter points ----------------
sns.stripplot(
    data=df,
    x="Genotype",
    y="Value",
    order=order,
    palette=palette,
    size=4,
    jitter=0.12,
    edgecolor=None,
    linewidth=0,
    alpha=1,
    ax=ax
)

# ---------------- dynamic Y scaling ----------------
y_min = df["Value"].min()
y_max = df["Value"].max()
range_y = y_max - y_min

# ---------------- significance bar ----------------
y = y_max + 0.25 * range_y
h = 0.12 * range_y

ax.plot([0,0,1,1], [y, y+h, y+h, y], lw=0.8, color="black")
ax.text(0.5, y+h*1.2, "**", ha="center", va="bottom", fontsize=8)

# ---------------- labels ----------------
ax.set_ylabel("Symptomatic leaves (%)", fontsize=8)
ax.set_xlabel("")

ax.tick_params(axis='both', labelsize=7)

sns.despine()

ax.set_ylim(3.5, y + 0.35*range_y)

plt.tight_layout()

plt.savefig(
    "C:/Users/tamer/Documents/PhD/ML/HLB_phenotype/leaf_phenotype.pdf",
    dpi=600,
    bbox_inches="tight"
)

plt.savefig(
    "C:/Users/tamer/Documents/PhD/ML/HLB_phenotype/leaf_phenotype.tif",
    dpi=600,
    bbox_inches="tight"
)

plt.show()
```
