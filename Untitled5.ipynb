```python
from scipy.stats import fisher_exact

# Tableau de contingence
#         Top50   NotTop50
# IL        42      127
# IO         8      113

table = [[9, 4963],
         [41, 5130]]

# Test d'enrichissement de IL dans le Top50
oddsratio, pvalue = fisher_exact(table, alternative='greater')

print("Table:", table)
print("Odds ratio:", oddsratio)
print("P-value:", pvalue)
```

```python
from scipy.stats import fisher_exact

table = [[47, 517],
         [77, 843]]

oratio, p_two = fisher_exact(table, alternative='two-sided')
_, p_greater = fisher_exact(table, alternative='greater')
_, p_less = fisher_exact(table, alternative='less')

print("OR =", oratio)
print("two-sided =", p_two)
print("greater =", p_greater)
print("less =", p_less)
```
