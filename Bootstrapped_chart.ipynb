```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# --------------------------------------------------
# GLOBAL GRAPHIC PARAMETERS
# --------------------------------------------------
plt.rcParams.update({
    "font.family": "Arial",
    "font.size": 8,
    "axes.labelsize": 9,
    "xtick.labelsize": 8,
    "ytick.labelsize": 8,
    "legend.fontsize": 8,
    "pdf.fonttype": 42,
    "ps.fonttype": 42
})

# --------------------------------------------------
# LOAD DATA
# --------------------------------------------------
df = pd.read_excel(
    'C:/Users/tamer/Documents/PhD/ML/ML_coefs/Coefs_BoLASSO.xlsx',
    sheet_name='SPSN_real'
)

# OPTIONAL: sort for readability
df = df.sort_values(
    "Percentage of selection in bootstrapped LASSO",
    ascending=True
)

# --------------------------------------------------
# COLOR MAP (CONSISTENT WITH PCA)
# --------------------------------------------------
color_map = {
    "Healthy orange": "#A6BDD7",
    "Infected orange": "#DD8452"
}

colors = df["Class"].map(color_map)

# --------------------------------------------------
# FIGURE SIZE (DOUBLE COLUMN SUPP FIG)
# --------------------------------------------------
fig, ax = plt.subplots(figsize=(7.1, 9))

# --------------------------------------------------
# BARPLOT
# --------------------------------------------------
ax.barh(
    df["Metabolite"],
    df["Percentage of selection in bootstrapped LASSO"],
    color=colors,
    edgecolor="black",
    linewidth=0.6,
    height=0.7
)

# --------------------------------------------------
# AXES STYLE
# --------------------------------------------------
for spine in ax.spines.values():
    spine.set_visible(True)
    spine.set_linewidth(0.8)

ax.tick_params(direction="out", length=3, width=0.8)

# subtle grid
ax.grid(
    axis="x",
    linestyle="--",
    linewidth=0.5,
    alpha=0.3
)

# --------------------------------------------------
# LABELS
# --------------------------------------------------
ax.set_xlabel("Selection frequency (% of bootstraps)")
ax.set_ylabel("")

ax.set_title(
    "Stable metabolites selected by bootstrapped LASSO",
    pad=8
)

ax.set_xlim(0, 100)

# --------------------------------------------------
# LEGEND
# --------------------------------------------------
legend_handles = [
    mpatches.Patch(
        facecolor="#A6BDD7",
        edgecolor="black",
        label="Healthy orange"
    ),
    mpatches.Patch(
        facecolor="#DD8452",
        edgecolor="black",
        label="Infected orange"
    )
]

ax.legend(
    handles=legend_handles,
    frameon=False,
    loc="lower right"
)

# --------------------------------------------------
# LAYOUT
# --------------------------------------------------
plt.tight_layout()

# --------------------------------------------------
# EXPORT
# --------------------------------------------------
plt.savefig(
    "C:/Users/tamer/Documents/PhD/ML/ML_coefs/BS_chart.pdf",
    bbox_inches="tight",
    transparent=True
)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# ==================================================
# FUNCTION
# ==================================================
def plot_bolasso(ax, sheet, color_map, xmax, panel_title):

    df = pd.read_excel(
        'C:/Users/tamer/Documents/PhD/ML/ML_coefs/Coefs_BoLASSO.xlsx',
        sheet_name=sheet
    )

    df = df.sort_values(
        "Percentage of selection in bootstrapped LASSO",
        ascending=True
    )

    colors = df["Class"].map(color_map)

    ax.barh(
        df["Metabolite"],
        df["Percentage of selection in bootstrapped LASSO"],
        color=colors,
        edgecolor="black",
        linewidth=0.6,
        height=0.7
    )

    for spine in ax.spines.values():
        spine.set_visible(True)
        spine.set_linewidth(0.8)

    ax.tick_params(direction="out", length=3, width=0.8)

    ax.grid(axis="x", linestyle="--", linewidth=0.5, alpha=0.3)

    ax.set_xlim(0, xmax)
    ax.set_xlabel("Selection frequency (%)")
    ax.set_title(panel_title, pad=6)

    ax.margins(y=0.01)


# ==================================================
# GLOBAL STYLE
# ==================================================
plt.rcParams.update({
    "font.family": "Arial",
    "font.size": 8,
    "axes.labelsize": 9,
    "xtick.labelsize": 7,
    "ytick.labelsize": 7,
    "legend.fontsize": 8,
    "pdf.fonttype": 42
})

# ==================================================
# FIGURE 2x2 GRID
# ==================================================
fig, axes = plt.subplots(
    nrows=2,
    ncols=2,
    figsize=(14, 12)
)

axes = axes.flatten()

# ==================================================
# PANELS
# ==================================================
plot_bolasso(
    axes[0],
    "LNSN_real",
    {
        "Healthy lemon": "#A6BDD7",
        "Healthy orange": "#F2B38F"
    },
    50,
    "Healthy lemon vs healthy orange"
)

plot_bolasso(
    axes[1],
    "LPSP_real",
    {
        "Infected lemon": "#4C72B0",
        "Infected orange": "#DD8452"
    },
    60,
    "Infected lemon vs infected orange"
)

plot_bolasso(
    axes[2],
    "LPLN_real",
    {
        "Healthy lemon": "#A6BDD7",
        "Infected lemon": "#4C72B0"
    },
    80,
    "Healthy lemon vs infected lemon"
)

plot_bolasso(
    axes[3],
    "SPSN_real",
    {
        "Healthy orange": "#F2B38F",
        "Infected orange": "#DD8452"
    },
    100,
    "Healthy orange vs infected orange"
)

# ==================================================
# PANEL LABELS
# ==================================================
labels = ["A", "B", "C", "D"]

for ax, label in zip(axes, labels):
    ax.text(
        -0.12,
        1.03,
        label,
        transform=ax.transAxes,
        fontsize=13,
        fontweight="bold"
    )

# ==================================================
# GLOBAL LEGEND
# ==================================================
legend_handles = [
    mpatches.Patch(facecolor="#A6BDD7", edgecolor="black", label="Healthy lemon"),
    mpatches.Patch(facecolor="#4C72B0", edgecolor="black", label="Infected lemon"),
    mpatches.Patch(facecolor="#F2B38F", edgecolor="black", label="Healthy orange"),
    mpatches.Patch(facecolor="#DD8452", edgecolor="black", label="Infected orange"),
]

fig.legend(
    handles=legend_handles,
    loc="lower center",
    ncol=4,
    frameon=False,
    bbox_to_anchor=(0.5, 0.02)
)

# ==================================================
# FINAL LAYOUT
# ==================================================
plt.tight_layout(rect=[0, 0.05, 1, 1])

# ==================================================
# EXPORT
# ==================================================
plt.savefig(
    "C:/Users/tamer/Documents/PhD/ML/ML_coefs/Figure_BoLASSO_ALL.pdf",
    bbox_inches="tight",
    transparent=True
)

plt.savefig(
    "C:/Users/tamer/Documents/PhD/ML/ML_coefs/Figure_BoLASSO_ALL.tif",
    bbox_inches="tight",
    transparent=True,
    format = 'tiff'
)

plt.show()
```
