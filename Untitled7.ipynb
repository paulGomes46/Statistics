```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# =========================
# PARAMETRES A MODIFIER
# =========================
file_path = "C:/Users/tamer/Documents/PhD/ML/saqib.xlsx"   # ou "my_dataset.xlsx"
sheet_name = "Sheet1"                 # seulement si Excel
target = 0.99                  # seuil de variance cumulée
metadata_cols = ["SampleID", "Class"]   # colonnes à exclure si présentes

# =========================
# LECTURE DU FICHIER
# =========================
if file_path.endswith(".csv"):
    df = pd.read_csv(file_path)
elif file_path.endswith(".xlsx") or file_path.endswith(".xls"):
    df = pd.read_excel(file_path, sheet_name=sheet_name)
else:
    raise ValueError("Format non supporté. Utilise un fichier CSV ou Excel.")

# =========================
# SELECTION DES VARIABLES NUMERIQUES
# =========================
# On enlève les colonnes metadata si elles existent
cols_to_drop = [col for col in metadata_cols if col in df.columns]
df_data = df.drop(columns=cols_to_drop, errors="ignore")

# Garder uniquement les colonnes numériques
df_data = df_data.select_dtypes(include=[np.number])

if df_data.shape[1] == 0:
    raise ValueError("Aucune colonne numérique trouvée dans le dataset.")

# =========================
# OPTIONNEL : gérer les valeurs manquantes
# =========================
# Ici on enlève les variables contenant au moins une NA
df_data = df_data.dropna(axis=1)

if df_data.shape[1] == 0:
    raise ValueError("Toutes les variables ont été supprimées à cause des NA.")

# =========================
# CALCUL DES VARIANCES
# =========================
# Variance de chaque variable à travers les samples
variances = df_data.var(axis=0, ddof=1)

# Trier par variance décroissante
variances_sorted = variances.sort_values(ascending=False)

# Variance totale
total_variance = variances_sorted.sum()

# Proportion cumulée de variance expliquée
cumulative_variance = variances_sorted.cumsum() / total_variance

# Nombre minimal de variables pour atteindre le seuil
n_features_target = np.argmax(cumulative_variance.values >= target) + 1
target_value = cumulative_variance.iloc[n_features_target - 1]

print(f"Nombre total de variables : {df_data.shape[1]}")
print(f"Nombre de variables pour atteindre {target*100:.0f}% de la variance : {n_features_target}")
print(f"Variance cumulée atteinte : {target_value:.4f}")

# =========================
# SAUVEGARDE DES RESULTATS
# =========================
results = pd.DataFrame({
    "Variable": variances_sorted.index,
    "Variance": variances_sorted.values,
    "Cumulative_variance_explained": cumulative_variance.values
})

results.to_csv("variance_ranking_results.csv", index=False)

# =========================
# PLOT
# =========================
plt.figure(figsize=(6, 5))
x = np.arange(1, len(cumulative_variance) + 1)

plt.plot(x, cumulative_variance.values, linewidth=2)
plt.axhline(y=target, linestyle="--")
plt.axvline(x=n_features_target, linestyle="--")
plt.scatter(n_features_target, target_value, zorder=3)

plt.xlabel("Variables (triées par variance)")
plt.ylabel("Variance cumulée expliquée")
plt.ylim(0, 1.02)
plt.xlim(1, len(cumulative_variance))
plt.tight_layout()
plt.savefig("cumulative_variance_plot.png", dpi=300)
plt.show()
```
