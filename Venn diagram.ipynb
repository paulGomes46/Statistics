```python
dataset_total_path = "C:/Users/tamer/Documents/PhD/ML/Venn/Venn_diagram.xlsx"
df = pd.read_excel(dataset_total_path, sheet_name = 'Pos')
df = df.set_index('Metabolites')

# Ensure the indicator columns are boolean
indicator_cols = df.columns.tolist()
df[indicator_cols] = df[indicator_cols].astype(bool)

df = df[df.any(axis=1)]

# Build upset data structure from indicator columns
upset_data = from_indicators(indicator_cols, data=df)

# Plot
plt.figure(figsize=(10, 5))
up = UpSet(
    upset_data,
    subset_size="count",      # intersection bar heights are counts
    show_counts=True,
    show_percentages=False,
    sort_by="cardinality",     # largest intersections first
    facecolor="black"
    ).plot()

plt.suptitle("UpSet plot of differential metabolites across comparisons", y=1.02)
plt.tight_layout()

plt.savefig("C:/Users/tamer/Documents/PhD/ML/Venn/Venn_pos.pdf", format="pdf", bbox_inches="tight")   
plt.savefig("C:/Users/tamer/Documents/PhD/ML/Venn/Venn_pos.png", dpi=600, bbox_inches="tight") 

plt.show()
```
