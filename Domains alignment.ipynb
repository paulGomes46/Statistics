```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Patch

# ===============================
# DATA (Pfam uniquement)
# ===============================
proteins = [
    {
        "name": "At2g20340",
        "length": 490,
        "domains": [
            {"start": 46, "end": 419, "type": "PF00282"},
        ]
    },
    {
        "name": "Cs5g_pb032480",
        "length": 491,
        "domains": [
            {"start": 43, "end": 416, "type": "PF00282"},
        ]
    },
    {
        "name": "Cs2g_pb002830",
        "length": 506,
        "domains": [
            {"start": 55, "end": 428, "type": "PF00282"},
        ]
    },
    {
        "name": "Cs3g_pb017900",
        "length": 506,
        "domains": [
            {"start": 54, "end": 411, "type": "PF00282"},
        ]
    }
]


# ===============================
# COLORS (plus contrastées)
# ===============================
color_map = {
    "PF00282": "#1b9e77",  # bleu clair plus net
}

# ===============================
# STYLE GLOBAL
# ===============================
plt.rcParams.update({
    "font.family": "Arial",
    "axes.linewidth": 1,
})

# ===============================
# PLOT
# ===============================
fig, ax = plt.subplots(figsize=(7, 3))

for i, prot in enumerate(proteins):
    y = i

    # backbone (plus discret)
    ax.plot(
        [0, prot["length"]],
        [y, y],
        color="black",
        linewidth=1.8,
        zorder=1
    )

    # domaines (plus épais et visibles)
    for dom in prot["domains"]:
        ax.add_patch(
            Rectangle(
                (dom["start"], y - 0.4),
                dom["end"] - dom["start"],
                0.8,
                facecolor=color_map[dom["type"]],
                edgecolor="black",
                linewidth=1,
                zorder=2
            )
        )

# ===============================
# FORMAT
# ===============================
ax.set_yticks(range(len(proteins)))
ax.set_yticklabels([p["name"] for p in proteins], fontsize=9)

ax.set_xlabel("Amino acid position (aa)", fontsize=10)

# limites propres
ax.set_xlim(0, max(p["length"] for p in proteins) + 10)
ax.set_ylim(-0.8, len(proteins) - 0.2)

# ticks plus visibles
ax.tick_params(axis='x', labelsize=9, length=4, width=1)

# clean
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

# ===============================
# LEGEND (plus propre)
# ===============================
legend_elements = [
    Patch(facecolor="#1b9e77", edgecolor='black', label='pyridoxal-dependent decarboxylase domain (PF00282)'),
]

ax.legend(
    handles=legend_elements,
    fontsize=8,
    frameon=False,
    loc="lower center",
    bbox_to_anchor=(0.5, 1.15),
    ncol=2
)

plt.tight_layout()

# ===============================
# EXPORT
# ===============================
plt.savefig("C:/Users/tamer/Documents/PhD/ML/Protein_domains/PAAS.pdf")
plt.savefig("C:/Users/tamer/Documents/PhD/ML/Protein_domains/PAAS.tiff", dpi=600)

plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Patch

# ===============================
# DATA (filtré proprement)
# ===============================
proteins = [
    {
        "name": "Cs4g_pb002040",
        "length": 454,
        "domains": [
            {"start": 1, "end": 24, "type": "SIGNAL"},
            {"start": 258, "end": 426, "type": "PF00201"},
        ]
    },
    {
        "name": "AT5G17050",
        "length": 390,
        "domains": [
            {"start": 1, "end": 24, "type": "SIGNAL"},
            {"start": 191, "end": 361, "type": "PF00201"},
        ]
    }
]

# ===============================
# COLORS (nouvelle palette)
# ===============================
color_map = {
    "SIGNAL": "#f28e2b",   # orange (signal peptide)
    "PF00201": "#59a14f",  # vert (UGT)
}

# ===============================
# STYLE GLOBAL
# ===============================
plt.rcParams.update({
    "font.family": "Arial",
    "axes.linewidth": 1,
})

# ===============================
# PLOT
# ===============================
fig, ax = plt.subplots(figsize=(7, 2.2))

for i, prot in enumerate(proteins):
    y = i

    # backbone
    ax.plot(
        [0, prot["length"]],
        [y, y],
        color="black",
        linewidth=1.8,
        zorder=1
    )

    # domains
    for dom in prot["domains"]:
        ax.add_patch(
            Rectangle(
                (dom["start"], y - 0.4),
                dom["end"] - dom["start"],
                0.8,
                facecolor=color_map[dom["type"]],
                edgecolor="black",
                linewidth=1,
                zorder=2
            )
        )

# ===============================
# FORMAT
# ===============================
ax.set_yticks(range(len(proteins)))
ax.set_yticklabels([p["name"] for p in proteins], fontsize=9)

ax.set_xlabel("Amino acid position (aa)", fontsize=10)

ax.set_xlim(0, max(p["length"] for p in proteins) + 10)
ax.set_ylim(-0.6, len(proteins) - 0.4)

ax.tick_params(axis='x', labelsize=9, length=4, width=1)

# clean
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

# ===============================
# LEGEND
# ===============================
legend_elements = [
    Patch(facecolor="#f28e2b", edgecolor='black', label='Signal peptide'),
    Patch(facecolor="#59a14f", edgecolor='black', label='UDP-glycosyltransferase domain (PF00201)')
]

ax.legend(
    handles=legend_elements,
    fontsize=8,
    frameon=False,
    loc="lower center",
    bbox_to_anchor=(0.5, 1.15),
    ncol=2
)

plt.tight_layout(rect=[0, 0, 1, 0.9])

# ===============================
# EXPORT
# ===============================
plt.savefig("C:/Users/tamer/Documents/PhD/ML/Protein_domains/F3G.pdf")
plt.savefig("C:/Users/tamer/Documents/PhD/ML/Protein_domains/F3G.tiff", dpi=600)

plt.show()
```
