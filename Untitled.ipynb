```python
dataset_path = "C:/Users/tamer/Documents/PhD/ML/metabolome_filtré.xlsx"
df_total = pd.read_excel(dataset_path)
df_total.shape
```

```python
df = df_total[(df_total["Class"] == 'LP') | (df_total["Class"] == 'SP')]
df['Malonylgenistin;']
df.head()
```

```python
X_total = df.drop('Class', axis=1)
y_total = df['Class']
trainset, testset = train_test_split(df, test_size = 0.3, random_state=1)
```

```python
def encodage(df):
    code = {
    'LP' : 1,
    'SP' : 0,
    'LN' : 3,
    'SN' : 4
}
# Appliquer ce dictionnaire aux colonnes du dataset, avec la fonction map
    for col in df.select_dtypes('object'):
        df[col] = df[col].map(code)

    return df


def preprocessing(df):
    df = encodage(df)

    X = df.drop(['Class'], axis = 1)
    y = df['Class']

    # compter le nombre d'échantillons restants dans le dataset après avoir été inputé
    print(y.value_counts())

    return X, y
```

```python
X_train, y_train = preprocessing(trainset)
X_test, y_test = preprocessing(testset)
```

```python
log2_transformer = FunctionTransformer(lambda x: np.log2(x + 1))
model = make_pipeline(log2_transformer, 
                      StandardScaler(), 
                      SelectFpr(f_classif, alpha=0.05),
                      SelectFromModel(LogisticRegression(penalty="elasticnet", l1_ratio = 0.3, solver="saga", random_state=0), threshold=None), 
                      SVC(kernel='linear', random_state=0))


#SelectFromModel(LogisticRegression(penalty="elasticnet", solver="saga", l1_ratio=0.5, random_state=0), threshold=None), 


#model.fit(X_train, y_train)
model.fit(X_total, y_total)
#y_pred = model.predict(X_test)
```

```python
### Get the variables selected IF some variables were filtered in the pipeline

# Masks from each selector
mask_anova = model.named_steps['selectfpr'].get_support()
mask_sfm   = model.named_steps['selectfrommodel'].get_support()

# Names after ANOVA, then after SelectFromModel
cols_after_anova = X_train.columns[mask_anova]
final_cols = cols_after_anova[mask_sfm]

print("Number of features kept:", final_cols.size)
print("Selected metabolites:", list(final_cols))
```

```python
### Get the variables and their coefficients

# 1) Masks and names you already have
mask_anova = model.named_steps['selectfpr'].get_support()
mask_sfm   = model.named_steps['selectfrommodel'].get_support()
cols_after_anova = X_train.columns[mask_anova]
final_cols = cols_after_anova[mask_sfm]

# 2) Get ElasticNet (inside SelectFromModel) coefficients
lr = model.named_steps['selectfrommodel'].estimator_   # LogisticRegression(...)
en_coef_all = lr.coef_.ravel()                         # aligned to cols_after_anova

# 3) Build a Series/DataFrame with names and coefs
en_coef_series = pd.Series(en_coef_all, index=cols_after_anova)

# 4) Keep only the finally selected features (non-zero by SFM)
en_coef_final = en_coef_series.loc[final_cols].sort_values(key=np.abs, ascending=False)

en_coef_final.to_excel("C:/Users/tamer/Documents/PhD/ML/EN_coef.xlsx")

print(en_coef_final)
```
